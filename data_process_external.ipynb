```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem.MolStandardize import rdMolStandardize

# -----------------------------
# Functions for standardization
# -----------------------------

def remove_salts_standardize(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            return None

        # 1️⃣ Normalize functional groups
        normalizer = rdMolStandardize.Normalizer()
        mol = normalizer.normalize(mol)

        # 2️⃣ Remove salts / keep largest fragment
        lf = rdMolStandardize.LargestFragmentChooser()
        mol = lf.choose(mol)

        # 3️⃣ Neutralize charges
        uncharger = rdMolStandardize.Uncharger()
        mol = uncharger.uncharge(mol)

        # 4️⃣ Canonical tautomer (removes tautomer ambiguity)
        te = rdMolStandardize.TautomerEnumerator()
        mol = te.Canonicalize(mol)

        # Convert back to SMILES
        return Chem.MolToSmiles(mol)
    except:
        return None

# -----------------------------
# Load your CSV
# -----------------------------

df = pd.read_csv("human_deduplicated.csv")  # <-- change filename

# -----------------------------
# Standardize SMILES
# -----------------------------

df["standard_smiles"] = df["standard_smiles"].apply(remove_salts_standardize)

# Remove failed molecules
df = df[df["standard_smiles"].notnull()].copy()

# Save result
df.to_csv("human_standardized_output.csv", index=False)

print("Standardization complete → saved to standardized_output.csv")
```
